# matplotlib colour maps

Some colourmaps of matplotlib



cat_red[:, :, 0], cat_red[:, :, 1], cat_red[:, :, 2] selects the second channel of the image array.

In a standard RGB image:

0 = Red channel

1 = Green channel

2 = Blue channel

So, 1 refers to the Green channel.

```python
plt.imshow(cat_red[:, :, 0],cmap='Grays')
```

```python
plt.imshow(cat_red[:, :, 1],cmap='Grays')
```

```python
plt.imshow(cat_red[:, :, 2],cmap='PuBu')
```
